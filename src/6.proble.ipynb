```python
import pandas as pd
dfMatches=pd.read_csv('../public/matches.csv')

# // Find a player who has won the highest number of Player of the Match awards for each season

result=dfMatches.groupby(['season','player_of_match']).size().reset_index(name='Player of the match count')
top_players = result.loc[result.groupby('season')['Player of the match count'].idxmax()].reset_index(drop=True)
top_players.to_csv('../Output/6-problem.csv')
top_players
```
